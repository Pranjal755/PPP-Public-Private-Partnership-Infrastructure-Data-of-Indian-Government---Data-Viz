## Q2
### Display 5 lowest performing projects with their difference in Agreed Project Cost and Actual Project Costs.

```python
df_2 = df[['Project Name','Project Cost(As per agreement) in Rs Crore','Project Cost(Actual) in Rs Crore']]
df_2.head()
```

```python
df_2.info()
```

```python
df_2['Project Cost(As per agreement) in Rs Crore'] = df_2['Project Cost(As per agreement) in Rs Crore'].replace('Not Available',np.nan)
df_2['Project Cost(As per agreement) in Rs Crore'] = df_2['Project Cost(As per agreement) in Rs Crore'].str.replace(',','')
df_2['Project Cost(As per agreement) in Rs Crore'] = df_2['Project Cost(As per agreement) in Rs Crore'].astype(float)
df_2['Project Cost(Actual) in Rs Crore'] = df_2['Project Cost(Actual) in Rs Crore'].replace('Not Available',np.nan)
df_2['Project Cost(Actual) in Rs Crore'] = df_2['Project Cost(Actual) in Rs Crore'].replace('Not Applicable',np.nan)
df_2['Project Cost(Actual) in Rs Crore'] = df_2['Project Cost(Actual) in Rs Crore'].str.replace(',','')
df_2['Project Cost(Actual) in Rs Crore'] = df_2['Project Cost(Actual) in Rs Crore'].astype(float)
df_2.info()
```

```python
df_2['Difference'] = df_2['Project Cost(As per agreement) in Rs Crore'] - df_2['Project Cost(Actual) in Rs Crore']
df_2.head()
```

```python
df_2.dropna(inplace=True)
```

```python
df_2.head()
```

```python
df_2.shape
```

```python
df_2.sort_values(by='Difference', ascending=False).head(5)
```

```python
df_2.sort_values(by='Difference')[:5]
```

```python
df_2_viz=df_2.sort_values(by='Difference')[:5]
df_2_viz['Difference'] = abs(df_2_viz['Difference'])
df_2_viz
```

```python
df_2_viz.info()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.barh(data=df_2_viz,width='Difference',y='Project Name')
plt.title('5 Lowest Performing Projects',loc='center',fontdict={'family':'serif','size':25})
plt.xlabel('Project Extra Costs',fontdict={'size':18})
plt.xticks(np.arange(0,5000,1000),fontsize=14)
plt.ylabel('Project Name',fontdict={'size':18})
plt.yticks(fontsize=16)
for i, v in enumerate(df_2_viz['Difference']):
    ax.text(v+25, i, int(v), color='black', fontsize=14, ha='left', va='center')
plt.show()
```

## Q6
### Display number of years in difference between Construction Completion Date (As per agreement) and Construction Completion Date (Actual).

```python
df_6 = df[['Project Name','Construction Completion Date(As per agreement)','Construction Completion Date(Actual)']]
df_6
```

```python
df_6['Construction Completion Date(As per agreement)'] = df_6['Construction Completion Date(As per agreement)'].replace('Not Available',np.nan)
df_6['Construction Completion Date(As per agreement)'] = df_6['Construction Completion Date(As per agreement)'].replace('Not Applicable',np.nan)
df_6['Construction Completion Date(Actual)'] = df_6['Construction Completion Date(Actual)'].replace('Not Available',np.nan)
df_6['Construction Completion Date(Actual)'] = df_6['Construction Completion Date(Actual)'].replace('Not Applicable',np.nan)
df_6.head()
```

```python
df_6.dropna(inplace=True)
```

```python
df_6.head()
```

```python
df_6.shape
```

```python
df_6['Construction Completion Date(As per agreement)'] = pd.to_datetime(df_6['Construction Completion Date(As per agreement)'])
df_6['Construction Completion Date(Actual)'] = pd.to_datetime(df_6['Construction Completion Date(Actual)'])
df_6['Difference'] = df_6['Construction Completion Date(Actual)'] - df_6['Construction Completion Date(As per agreement)']
df_6.head()
```

```python
df_6.sort_values(by='Difference', ascending=False).head(5)
```

```python
df_6['Difference'] = df_6['Difference'].astype(str)
diff = []
for i in df_6['Difference'].values:
  diff.append(i.split(' ')[0])
df_6['Difference'] = diff
df_6['Difference'] = df_6['Difference'].astype(int)
df_6.info()
```

```python
df_6['Difference'] = df_6['Difference']/365
df_6 = df_6.sort_values(by='Difference',ascending=False)[:5]
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.barh(data=df_6,width='Difference',y='Project Name')
plt.title('5 Longest Projects',loc='center',fontdict={'family':'serif','size':25})
plt.xlabel('Extra Duration',fontdict={'size':18})
plt.xticks(np.arange(0,10,1),fontsize=14)
plt.ylabel('Project Name',fontdict={'size':18})
plt.yticks(fontsize=16)
for i, v in enumerate(df_6['Difference']):
    ax.text(v+0.1, i, str(int(v))+' years, '+str(int((v-int(v))*12))+' months', color='black', fontsize=14, ha='left', va='center')
plt.show()
```
